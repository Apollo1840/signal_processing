```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_devil_func(time_interval):
    sample_fs = 10
    n_samples = time_interval * sample_fs 
    # n_sampels = n_samples + 1 
    t = np.linspace(0.0, time_interval, n_samples)

    w1 = 5
    w2 = 8
    y = np.cos(w1 * 2.0*np.pi*t) + 0.5*np.cos(w2 * 2.0*np.pi*t)

    plt.plot(t, y)
    plt.show()
    
    
```

```python
for t in range(1, 6):
    plot_devil_func(t)
```
